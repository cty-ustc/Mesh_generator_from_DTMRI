```python
mat1 = scipy.io.loadmat('DTI080803/e1.mat')
e1 = mat1['e1']
mat2 = scipy.io.loadmat('DTI080803/e2.mat')
e2 = mat2['e2']
mat3 = scipy.io.loadmat('DTI080803/e3.mat')
e3 = mat3['e3']
mat11 = scipy.io.loadmat('DTI080803/v11.mat')
b11 = mat11['v11']
mat12 = scipy.io.loadmat('DTI080803/v12.mat')   
b12 = mat12['v12']
mat13 = scipy.io.loadmat('DTI080803/v13.mat')
b13 = mat13['v13']

import scipy.io
from sklearn.neighbors import KDTree
import pickle
import os
from vtk import *
from vtk.util.numpy_support import vtk_to_numpy
from sklearn.neighbors import KDTree
import logging





each_x_y = 1
each_z = 1




#прореживание
e1 = e1[::each_x_y][:,::each_x_y][:,:,::each_z]
e2 = e2[::each_x_y][:,::each_x_y][:,:,::each_z]
e3 = e3[::each_x_y][:,::each_x_y][:,:,::each_z]
b11 = b11[::each_x_y][:,::each_x_y][:,:,::each_z]
b12 = b12[::each_x_y][:,::each_x_y][:,:,::each_z]
b13 = b13[::each_x_y][:,::each_x_y][:,:,::each_z]


len1 = len(b11)
len2 = len(b11[0])
len3 = len(b11[0][0])
len_ = len1*len2*len3

results_points = np.array([])
results_vectors = np.array([])

MASK_THRESHOLD = 3.5 * 10 ** (-6)

multiply_x = 0.3*0.1 # Коэфициент переводящий в см
multiply_y = 0.3*0.1
multiply_z = 0.8*0.1
valuable = 0

for i in xrange(0, len1):
        for j in xrange(0, len2):
            for k in xrange(0, len3):
                if e3[i][j][k] >= MASK_THRESHOLD:
                    valuable += 1

with open(os.path.join('result/surface02.vtk'), 'w') as out_mesh:

    out_mesh.write("# vtk DataFile Version 2.0\n")
    out_mesh.write("Really cool data\n")
    out_mesh.write("ASCII\n")
    out_mesh.write("DATASET UNSTRUCTURED_GRID\n")
    out_mesh.write("POINTS {pcount} double\n".format(pcount=valuable))
    for i in xrange(0, len1):
        for j in xrange(0, len2):
            for k in xrange(0, len3):
                if e3[i][j][k] >= MASK_THRESHOLD:
                    out_mesh.write("{0} {1} {2}\n".format(multiply_x*each_x_y*i, multiply_y*each_x_y*j, multiply_z*each_z*k))

    out_mesh.write("CELLS {pcount} {pcount2}\n".format(pcount=valuable, pcount2=valuable*2))
    for j in xrange(valuable):
            out_mesh.write("1 {0}\n".format(j))
    out_mesh.write("CELL_TYPES {pcount}\n".format(pcount=valuable))
    for k in xrange(valuable):
            out_mesh.write("1\n")
    out_mesh.write("POINT_DATA {pcount}\n".format(pcount=valuable))
    out_mesh.write("VECTORS DTMRIFiberOrientation double\n")
    for i in xrange(0, len1):
        for j in xrange(0, len2):
            for k in xrange(0, len3):
                if e3[i][j][k] >= MASK_THRESHOLD:
                    out_mesh.write("{0} {1} {2}\n".format(b11[i][j][k], b12[i][j][k], b13[i][j][k]))

```

```python
def process(image, n):
    val = filters.threshold_otsu(image)
    mask = image < val
    im2 = np.logical_not(mask)
    im16 = remove_small_holes(im2, 300, connectivity=8)
    im17 = remove_small_objects(im16, 6)
    im3 = erosion(im17, square(4))
    im18 = remove_small_objects(im3, 500)
    if n >= 90:
        im9 = dilation(im18, square(4))
        im10 = np.logical_not(im9)
        return im10
    else:      
        
        im6 = erosion(im18, square(2))
        #im8 = mean(im6, disk(2))
        im9 = dilation(im6, square(2))
        #im5 = dilation(im6, square(3))
        #im8 = mean(im5, disk(1))
        #im10 = dilation(im8, square(5))
        #im11 = erosion(im10, square(5))
        #im15 = remove_small_holes(im11, 55, connectivity=8)
        #io.imshow(im15)
        im16 = np.logical_not(im9)   
        return im16
clean_list = np.zeros((len(e3[:,1,1]), len(e3[1,:,1]), len(e3[1,1,:])))
#clean_array = np.append(clean_array,e3[:,:,1], axis=0)
print e3[:,:,129]
print len(e3[0][0])
n = 0
for i in xrange(len(e3[1,1,:])):
    clean_list[:,:,i] = process(e3[:,:,i], n)
    n += 1
    
    
```
